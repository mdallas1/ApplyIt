Exercise 1: Write a program to compare the standard Euler method and the Euler-Comer method applied to a particle travling in one dimension. First assume the force $F(x) = -kx-bx^3$.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Define force
k = 1
b = 4
F = lambda x: -k*x-b*(x**3)
```

```python
#=============================
#Euler Method
#=============================

#Initialize data
x_e = []
v_e = []
t_e = []

#Set Initial Conditions
x_e.append(1)
v_e.append(0)
t_e.append(0)
```

```python
#Define step length
d = 0.05

#Write loop implementing Euler Method
for i in range(0,121):
  f = F(x_e[i])
  x_new = x_e[i]+v_e[i]*d
  v_new = v_e[i]+f*d
  x_e.append(x_new)
  v_e.append(v_new)
  t_e.append(t_e[i]+d)
```

```python
#Plot the velocity vs position
plt.figure(1)
plt.plot(x_e,v_e,'r*')
plt.xlabel('x')
plt.ylabel('v')
plt.title('Euler')
plt.figure(2)
plt.plot(t_e,x_e,'k',linewidth=3)
plt.xlabel('t')
plt.ylabel('x')

plt.tight_layout(pad=1.5)
plt.show()
```

```python
#=============================
#Euler-Cromer Method
#=============================

#Initialize data
x_ec = []
v_ec = []
t_ec = []

#Set Initial Conditions
x_ec.append(1)
v_ec.append(0)
t_ec.append(0)
```

```python
#Define step length
d = 0.05

#Write loop implementing Euler-Cromer Method
for i in range(0,121):
  f = F(x_ec[i])
  v_new = v_ec[i]+f*d
  x_new = x_ec[i]+v_new*d
  x_ec.append(x_new)
  v_ec.append(v_new)
  t_ec.append(t_ec[i]+d)
```

```python
#Plot the velocity vs position
plt.figure(3)
plt.plot(x_ec,v_ec,'bo')
plt.title('Euler-Cromer')
plt.xlabel('x')
plt.ylabel('v')
plt.title('Euler-Cromer')
plt.grid('on')
plt.figure(4)
plt.plot(t_ec,x_ec,'k',linewidth=3)
plt.xlabel('t')
plt.ylabel('x')
plt.grid('on')

plt.tight_layout
plt.show()
```
